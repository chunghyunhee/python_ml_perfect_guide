## 2. DataFrame과 리스트, 딕셔너리, 넘파이, ndarray 상호변환

기본적으로 dataframe은 파이썬의 리스트, 딕셔너리, numpy의 ndarray 등 다양한 데이터로부터 생성이 가능하고, 이런 형태로 변환 또한 가능하다. 여기서 dataframe과 numpy의 ndarray의 상호간 변환은 매우 빈번하게 일어난다

#### (1) list,array,dictionary를 dataframe으로 변환하는 과정

````python
#넘파이 ndarray, 리스트, 딕셔너리를 dataframe으로 변환하기
#1. data : 리스트나 딕셔너리, 넘파이 ndarray값을 통해 전달받음
#2. columns : 칼럼명 리스트를 입력받아 쉽게 dataframe을 생성한다
import numpy as np
import pandas as pd


col_name1=['col1'] #칼럼명 지정하기 위한 리스트
list1=[1,2,3]      #리스트->dataframe으로 변환하기 위한 리스트 생성
array1=np.array(list1) #행렬 -> dataframe으로 변환하기 위한 리스트 생성
print("array1 shape :", array1.shape) #생성한 행렬의 형태 확인
````

````python
#리스트로 dataframe생성하는 방법
df_list1=pd.DataFrame(list1, columns=col_name1)
print("1차원 리스트로 만든 dataframe:\n",df_list1)
````

````python
#numpy ndarray로 dataframe 생성하는 방법
df_array1=pd.DataFrame(array1, columns=col_name1)
print("1차원 ndarray로 만든 dataframe:\n",df_array1)
````

앞에서는 1차원 형태의 데이터를 기반으로 dataframe을 생성했었다. 이제는 2차원 형태의 데이터를 기반으로 dataframe을 생성한다. 2행 3열 형태의 리스트와 ndarray를 기반으로 dataframe을 생성하므로 칼럼명은 3개가 필요하다

````python
#칼럼명 생성
col_name2=['col1','col2','col3']

#2행, 3열의 리스트와 ndarray를 생성
list2=[[1,2,3],[11,12,13]] #리스트
array2=np.array(list2) #행렬
print(array2) #생성한 행렬의 형태 확인
print(array2.shape)
````

````python
#리스트를 데이터프레임 형태로 변환
df_list2=pd.DataFrame(list2, columns=col_name2)
print("2차원 데이터로 만든 dataframe:\n",df_list2)
````

````python
#ndarray를 데이터프레임 형태로 변환
df_array2=pd.DataFrame(array2, columns=col_name2)
print("2차원 ndarray로 만든 dataframe:\n",df_array2)
````

````python
#딕셔너리를 데이터프레임 형태로 변환
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]} #key : value값으로 이루어져 있기 때문에 key->column, value->data형태로 mapping
df_dict=pd.DataFrame(dict)
print("딕셔너리로 만든 dataframe:\n",df_dict)
````

### 3. dataframe의 칼럼 데이터 세트 생성과 수정과정

````python
#데이터프레임에서 새로운 열 생성
#새롭게 생성한 열에 0을 일괄적으로 대입
titanic_df['Age_0']=0 
titanic_df.head() #확인
````

seiries에 상숫값을 할당하면 series의 모든 데이터 세트에 상숫값이 일괄적으로 적용된다. ndarray에 상숫값을 대입할때 같은 값이 일괄적으로 더해지는 것과 같은 원리 

````python
#기존 dataframe의 열에 있는 데이터를 활용하여 새로운 칼럼 생성
titanic_df['Age_by_10']=titanic_df['Age']*10
titanic_df.head()
````

````python
#기존 데이터 칼럼 값도 일괄적으로 연산할 수있다. 
titanic_df['Age_by_10']=titanic_df['Age_by_10']+100
titanic_df.head()
````

#### dataframe의 데이터 삭제

**데이터프레임이름.drop("칼럼명",axis=0, index=None, columns=None, level=None, inplace=False. errors='raise')**

(axis=0  :행삭제
axis=1  :열삭제)

여기서 drop을 적용했을 때, 원래 데이터프레임에 적용되지 않는다.(drop의 인자중 inplace=False가 디폴트값이기 때문이다) 따라서 분석시 계속 사용하기 위해서는 새로운 변수로 전달하여 사용한다

````python
titanic_drop_df=titanic_df.drop('Age_0',axis=1)
titanic_drop_df.head()
````

````python
#앞에서 생성한 열을 모두 삭제한다. 
drop_result=titanic_df.drop(['Age_0','Age_by_10'],axis=1, inplace=True) #inplace=True로 지정하는 경우,none값이 반환되고, 
                                                                         #원래 데이터프레임에 변경값이 적용된다
print('inplace=TRUE로 drop한 후 반환된 값:',drop_result)
titanic_df.head(3)
````

````python
#행을 삭제 (인덱스 값 0,1,2를 삭제하는 예제)
pd.set_option('display.width',1000)          #option
pd.set_option('display.max_colwidth',15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))                     #행을 삭제하기 이전

titanic_df.drop([0,1,2],axis=0, inplace=True) #행의 일부를 삭제
print("#### after axis 0 drop ####")
print(titanic_df.head(3))
````

(2) dataframe.ix[], dataframe.iloc(), dataframe.loc()

- dataframe.iloc[] : 컬럼 **위치**를 기반

````python
#DeprecationWarning경고 무시하기
#.ix를 사용하면 명칭기반 인덱싱과 위치기반 인덱싱이 헷갈리기 때문에 warning이 출력된다. 따라서 잘 안쓰는 경우이다.
#여기선 경고를 출력하지 않기위해....
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
````

명칭기반 인덱싱과 위치기반 인덱싱의 구분
1. 명칭기반 인덱싱 : 칼럼의 명칭을 기반으로 위치를 지정하는 경우

### unpacking

결국 unpacking은 여러가지의 값을 가진 하나의 변수를 여러변수로 나누는 것을 의미한다.
```Python 
from sklearn import svm

params = {
    'C': 1.0, 
    'kernel': 'precomputed', 
    'gamma' : 'auto_deprecated'
}

clf = svm.SVC(**params)
```

상기 코드는 'params' 라는 딕셔너리에 SVC 생성자 함수의 인자를 정의하고, **unpacking**연산자를 사용해 딕셔너리를 SVC 생성자 함수의 파라미터에 인자로 전달하는 코드.<br> 

**unpacking**연산자는 딕셔너리의 key는 파라미터 변수 이름, value는 인자로 치환하여 정확히는 **unpacking**하여 함수를 호출해 주는 연산자입니다. 

아래 코드는 위 params 딕셔너리가 unpacking되어 실제 실행되는 코드의 예시이다.
```Python 
from sklearn import svm
clf = svm.SVC(C=1.0, kernel='precomputed', gamma='auto_deprecated')
```

````python
params = (0,1,2,3,4)
def tuple_unpacking(first, second, *args): 
    print(first, second)
    print(args)
    
tuple_unpacking(*params)
````

````python
params = {'name':'hyunhee', 'gender':'female', 'age':10, 'isHansome':False}

def dict_unpacking(name, gender,*argv,  **kwargs):#argv는 튜플, kwargs는 키워드(딕셔너리), positional은 위치 선정 중요
    print(name, 'is',gender)
    print(kwargs)
    
dict_unpacking(**params)
````

**unpacking** 연산자를 사용하여 함수를 호출할 때의 장점은 코드가 가독성이 높아진다. 기본적으로 sklean, tensorflow 머신러닝 프레임워크같은 라이브러리들은 10개가 넘어가는 매개변수를 수정해 줘야하는 경우가 존재.. 그렇기 때문에 매개변수를 딕셔너리 리스트로 관리한다면 코드의 가독성이 올라갈 뿐 아니라 함수 호출하는 행위를 데이터로 관리 가능.

### map(f, iterable)
map(f, iterable)은 함수(f)와 반복 가능한(iterable) 자료형을 인자로 받습니다. map은 입력받은 자료형의 각 요소를 함수 f가 수행한 결과를 돌려주는 함수입니다.

````python
result = map(lambda x: 2*x, [1,2,3,4,5]) #인자를 함수로 전달
list(result)
````

map함수와 + unpacking연산자의 사용

````python
from random import sample

def sentence(s, v, o):
    return f'{s} {v} {o}'

pick = lambda list_ : sample(list_, 1).pop() #랜덤추출 1개

list_s = ['he', 'she', 'i', 'you', 'we']
list_v = ['practive', 'enjoy', 'love', 'like']
list_o = ['python', 'java', 'machine learning', 'deep learning']

params =[ 
    (pick(list_s),pick(list_v), pick(list_o))
    for _ in range(10) ]

list(map(lambda argv:sentence(*argv) , params)) 
#여러개의 parameter을 선언하기 위함.(map은 하나의 인자만 받음)
````

#### filter(f_condition, iterable )

filter란 무엇인가를 걸러낸다는 뜻으로 filter 함수도 동일한 의미!

filter 함수는 첫 번째 인수로 함수 이름을, 두 번째 인수로 그 함수에 차례로 들어갈 반복 가능한 자료형을 받는다. 그리고 두 번째 인수인 반복 가능한 자료형 요소가 첫 번째 인수인 함수에 입력되었을 때 반환 값이 참인 것만 걸러내 돌려준다

````python
list(filter(lambda x: 3>x, [1,2,3,4,5]))
````
